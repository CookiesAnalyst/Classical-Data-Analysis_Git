## Implement a Regression model with multiple variables: independent variables “Age” and “Fare”; and “Survived “ as dependent variable.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/Users/sergi/Desktop/MBDA/Master/Data Science Foundations/Data Science Foundations_Git/Session_2/2_titanic_dataset.csv')
```

```python
df.head()
```

```python
df = df[['Survived','Age', 'Fare']]
```

```python
df.head()
```

### With a Constant

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0)
```

```python
df.info()
```

```python
X = df[['Age', 'Fare']]
y = df['Survived']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```

### Without a Constant

```python
X = df[['Age', 'Fare']]
y = df['Survived']
```

```python
X = sm.add_constant(X)
X["const"] = X["const"].apply(lambda x: x*0)
model = sm.OLS(y, X).fit()
model.summary()
```

Age and Fare are not good predictors for surviving in the Titanic.

We have obtained bad models with linear regression. The R-Squared is 0.083 with a constant and 0.374 without a constant. Without a constant the model is so much better, but still really bad despite we have low p-values.
