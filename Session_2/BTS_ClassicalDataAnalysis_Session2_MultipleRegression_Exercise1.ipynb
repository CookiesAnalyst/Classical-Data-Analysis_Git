# 1. Data

Let's take a look again at some data - Advertising - , ask some questions about that data, and then use linear regression (with one and multiple variables) to answer those questions!

```python
# imports
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read data into a DataFrame
df = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
df.head()
```

```python
df
```

# 2 - Linear Regression with multiple variables in Statsmodels

## 2.1. Regression model with multiple variables and with a constant for Situation 1

> Put your code here

```python
X = df[["TV", "radio"]]
y = df["sales"]
```

```python
X = sm.add_constant(X) 
X.head()
```

```python
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
```

```python
predictions = model.predict(X) 
predictions
```

```python
model.summary()
```

## 2.2. Regression model with multiple variables and with a constant for Situation 2

> Put your code here

```python
X = df[["TV", "radio", "newspaper"]]
y = df["sales"]
```

```python
X = sm.add_constant(X) 
X.head()
```

```python
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
```

```python
predictions = model.predict(X) 
predictions
```

```python
model.summary()
```

## 2.3. Regression model with multiple variables and without a constant for Situation 1

> Put your code here

```python
X = df[["TV", "radio"]]
y = df["sales"]
```

```python
X = sm.add_constant(X) 
X.head()
```

```python
X["const"] = X["const"].apply(lambda x: x*0)
X
```

```python
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
```

```python
predictions = model.predict(X) 
predictions
```

```python
model.summary()
```

## 2.4. Regression model with multiple variables and without a constant for Situation 2

> Put your code here

```python
X = df[["TV", "radio", "newspaper"]]
y = df["sales"]
```

```python
X = sm.add_constant(X) 
X.head()
```

```python
X["const"] = X["const"].apply(lambda x: x*0)
X
```

```python
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
```

```python
predictions = model.predict(X) 
predictions
```

```python
model.summary()
```
